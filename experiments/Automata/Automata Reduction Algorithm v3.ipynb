```python
class automata1(object):
#     X;
#     E;
#     f;
#     Gamma;
#     X0;
    
    def __init__(self,X,E,f,Gamma,X0):
        self.X = X;
        self.E = E;
        self.f = f;
        self.Gamma = Gamma;
        self.X0 = X0;
    
    def define_index_fcn(self,I_fcn):
        self.I_func = I_fcn;
        print('Index Function defined.')
    
    def inf_mapping_in_string(self,s):
        #Apply the "projection" to the current word according to the index function. 
        
        if 'I_func' not in self.__dict__:
            print('Error. I_func is not defined yet.')
            return None
        
        if len(s) is 0:
            return None
        elif len(s) is 1:
            if self.I_func[s[0]] is 1:
                return s
            else:
                return self.I_func(None)
        else:
            if self.I_func[s[-1]] is 1:
                return self.inf_mapping(s[0:-1]) + [s[-1]]
            else:
                return self.inf_mapping(s[0:-1])
            
    
    def print0(self):
        print('X  = ',str(self.X))
        print('E  = ',str(self.E))
        print('f  = ',str(self.f))
        for key in self.Gamma:
            print('f(',str(key),')=',str(self.Gamma[key]))
            
        print('X0 = ',str(self.X0))
        
    def find_maximal_indist_set():
        print('Temp')

def diff(first, second):
        second = set(second)
        return [item for item in first if item not in second]
        
#Define an automaton according to W. Wang et. al from 2007 (Prof. LaFortune)        
E4 = ['q0','q1']
f4 = [('s0',E4[0],'s1'),('s1',E4[1],'s2'),('s2',E4[0],'s3'),('s3',E4[1],'s4'),('s4',E4[0],'s0')]
G4 = dict()
G4['s0'] = E4[0]
G4['s1'] = E4[1]
G4['s2'] = E4[0]
G4['s3'] = E4[1]
G4['s4'] = E4[0]

A4 = automata1(X2,E4,f4,G4,X2[0])
print('Automata1 instance created.')
A4.print0()

#Define an index function according to LaFortune
E4_o = ['q0']
E4_uo = diff(E4,E4_o)
I_func4 = dict()
for key in E4_o:
    I_func4[key] = 1
for key in E4_uo:
    I_func4[key] = 0

A4.define_index_fcn(I_func4)

#Define the theta projection

s1 = [E4[0],E4[1],E4[0]]
s2 = [E4[0],E4[1]]
s3 = [E4[0],E4[1],E4[0],E4[1],E4[0],E4[0]]
s_tests = (s1,s2,s3)
for string0 in s_tests:
    print('Projection Mapping of ' + str(string0))
    print(A4.inf_mapping(string0))
```

```python
class fsm0(object):
    #The 2017 Automatica Paper by De Santis et. al. describes a useful property for us in terms of trajectories
    #instead of just transitions. (Unobservable or indistinguishable.)
    #Needed for Initialization
    # - X
    # - X0
    # - Y
    # - H
    # - Delta
    
    def __init__(self,X,X0,Y,H,Delta):
        self.X  = X;
        self.X0 = X0;
        self.Y  = Y;
        self.H  = H;
        self.Delta = Delta;
    
    def print0(self):
        print('X  = ',str(self.X))
        print('X0 = ',str(self.X0))
        print('Y  = ',str(self.Y))
        for key in self.H:
            print('H(',str(key),')=',str(self.H[key]))
        
        print('Delta = ',str(self.Delta))
    
    def create_Pi(self):
        #Description:
        # According to this paper, we need to define special set Pi.
        Pi = []
        for i in self.X:
            for j in self.X:
                #Key Criterion
                if self.H[i] is self.H[j]:
                    Pi += [ (i,j) ]

        self.Pi = Pi
        return Pi #For the sake of debugging
    
    def create_Theta(self):
        #Description:
        # According to this paper, we need to define special set Pi.
        Theta = []
        for i in self.X:
            Theta += [ (i,i) ]

        self.Theta = Theta
        return Theta #For the sake of debugging
    
    def succ(self,i):
        #Description:
        # According to paper
        #  succ(i) = { j in X : (i,j) in Delta }
        
        s_states = []
        for ij in self.Delta:
            if ij[0] is i:
                s_states += [ij[1]]
                
        return s_states
    
    def pre(self,i):
        #Description:
        # According to paper
        #   pre(i) = { j in X : (j,i) in Delta }
        
        p_states = []
        for ji in self.Delta:
            if ji[1] is i:
                p_states += [ji[0]]
                
        return p_states
    
    def S_kp1(self,S_k):
        #Description:
        # Equation (1) in the 2017 text describes a recursion that we would like to apply.
        #
        #Assumes:
        # - Pi already defined.
        temp_Skp1 = []
        temp_cp = []
        for ij in self.Pi:
            temp_cp = cart_prod(self.pre(ij[0]),self.pre(ij[1]))
            if intersection(temp_cp,S_k) is not None:
                temp_Skp1 += [ij]
            
        return temp_Skp1;
            
    
    def find_maximal_indist_set():
        print('Temp')

def diff(first, second):
        second = set(second)
        return [item for item in first if item not in second]

def cart_prod(l1,l2):
    #Description:
    # The cartesian product of the symbols listed in list 1 (l1) and list 2 (l2).
    
    product0 = []
    for item1 in l1:
        for item2 in l2:
            product0 += [(l1,l2)]
    
    return product0

def intersection(lst1,lst2):
    return [value for value in lst1 if value in lst2]
    
#Create the fsm version for diagnosability/indistinguishability analysis.
print('Considering Automatica FSM 2017 Paper.')
X5 = X2
X5_As = [X5[0],X5[2],X5[4]]
X5_Ms = diff(X5,X5_As)

X0_5 = X5[0];

Y5 = ['0','1']

H5 = dict()
for x in X5:
    if x in X5_As:
        H5[x] = Y5[1]
    else:
        H5[x] = Y5[0]
        
D5 = [ (X5[0],X5[1]),(X5[1],X5[2]),(X5[2],X5[3]),(X5[3],X5[4]),(X5[4],X5[0]) ]
        
A5 = fsm0(X5,X0_5,Y5,H5,D5)
print('Created "Automaton" (really an FSM) according to this paper.')
A5.print0()

print('Pi(A5)=',A5.create_Pi())
print('Theta(A5)=',A5.create_Theta())
print('S_{k+1} = ',A5.S_kp1( [(X0[0],X0[0])] ))
```
