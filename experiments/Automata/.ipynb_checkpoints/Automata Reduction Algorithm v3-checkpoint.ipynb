# 'Observability of Finite State Systems: A Unifying Framework'

The title of this section is the title of a paper submitted to Automatica in 2017 by E. De Santis et. al. In the paper, an algorithm is presented that seems to do what we want to do (finding indistinguishable $\textit{executions}$ of a graph). The proofs use Finite State Machines instead of Automata, but I think the spirit can be used again. Their definition of a Finite State Machine in this work is the following:

$$ M = (X,X_0,Y,H,\Delta) $$

where
+ $X$ is the finite set of states
+ $X_0 \subset X$ is the set of initial states
+ $Y$ is the finite set of outputs
+ $H: X \mapsto Y$ is the output function
+ $\Delta \subset X \times X$ is the transition relation

For the given Finite State Machine, we can also derive the following sets and operations:
+ $ \Pi = \{ (i,j) \in X \times X : H(i) = H(j) \} $
+ $ \Theta = \{ (i,j) \in X \times X : i = j \} $
+ $ succ(i) = \{ j \in X : (i,j) \in \Delta \} $
+ $ pre(i) = \{ j \in X : (j,i) \in \Delta \} $

With all of the above ingredients, the algorithm must determine something that has been referred to currently as 'localizability.' The work of De Santis et. al. suggests that 'localizability' is a consequence of "backward Distinguishability" according to their definition.

## Backwards Distinguishability

Backwards Distinguishability is presented in Definition 7 of the above paper and is written as follows:
> Two state trajectories $\mathcal{x}_1$ and $\mathcal{x}_2$ in $\mathcal{X}^*$ are called $\underline{\text{indistinguishable}}$ if $\mathbf{y}(\mathcal{x}_1) = \mathbf{y}(\mathcal{x}_2)$.

> The pair $(i,j) \in \Pi$ is $\underline{k\text{-forward indistinguishable}}$ if there exist $\mathcal{x}_1 \in \mathcal{X}_{\{i \} }$ and $\mathcal{x}_2 \in \mathcal{X}_{\{j \} }$, such that $|\mathcal{x}_1|=|\mathcal{x}_2|=k$ and $\mathbf{y}(\mathcal{x}_1) = \mathbf{y}(\mathcal{x}_2)$.

> The pair $(i,j) \in \Sigma \subset \Pi$ is $\underline{k\text{-backward indistinguishable in } \Sigma}$ if there exist $\mathcal{x}_1 \in \mathcal{X}^{\{i \} }$ and $\mathcal{x}_2 \in \mathcal{X}^{\{j \} }$, such that $|\mathcal{x}_1|=|\mathcal{x}_2|=k$, $[\mathcal{x}_1(h) \in \Sigma, \mathcal{x}_2(h) \in \Sigma ] \; \forall h \in [1,k]$, and $\mathbf{y}(\mathcal{x}_1) = \mathbf{y}(\mathcal{x}_2)$.

where the following definitions were used:
+ $\mathcal{X}^*$ is the set of state executions of M
+ $\mathcal{X}_{\Psi}$ is the set of state executions $\mathcal{x} \in \mathcal{X}^*$ with $\mathcal{x}(1) \in \Psi$, ($\Psi \subset X$)
+ $\mathcal{X}_{\Psi,\infty}$ is the set of infinite state executions $\mathcal{x} \in \mathcal{X}^*$ with $\mathcal{x}(1) \in \Psi$, ($\Psi \subset X$)
+ $\mathcal{X}^{\Psi}$ is the set of finite state executions $\mathcal{x} \in \mathcal{X}^*$ with last symbol in $\Psi$, ($\Psi \subset X$)

With this in mind, I would like to the following proposition:

## Proposition

A automaton (or fsm) is localizable if for every state in the automata, there exists a maximum $k$ such that the all pairs of states $(x_1,x_2)$ are k backwards distinguishable. In math,

$$ \text{Localizable} \quad \iff (\forall x \in X)((\exists k^*(x)) \; x \text{ is k-backwards indistinguishable } \forall k < k^*(x) ) $$

The number $k^*$ can be approximated by adding 1 to the longest word/string in the set $B^*(\Sigma)$, also known as the the maximal set of pairs $(i,j) \in \Pi$ that are k-backward indistinguishable $\forall k \in \mathbb{Z}, \; k \geq 1$. $B^*(\Sigma)$ can be found using the following recursion:

1. Let $B_1(\Sigma) = \Sigma$
2. For $k=1,2,...$,
   $$ B_{k+1} = \left\{ (i,j) \in B_k(\Sigma) \; | \; (pre(i) \times pre(j)) \cap B_k(\Sigma) \neq \emptyset \right\} $$

```python
class fsm0(object):
    #The 2017 Automatica Paper by De Santis et. al. describes a useful property for us in terms of trajectories
    #instead of just transitions. (Unobservable or indistinguishable.)
    #Needed for Initialization
    # - X
    # - X0
    # - Y
    # - H
    # - Delta
    
    def __init__(self,X,X0,Y,H,Delta):
        self.X  = X;
        self.X0 = X0;
        self.Y  = Y;
        self.H  = H;
        self.Delta = Delta;
        
        self.create_Theta();
        self.create_Pi();
    
    def print0(self):
        print('X  = ',str(self.X))
        print('X0 = ',str(self.X0))
        print('Y  = ',str(self.Y))
        for key in self.H:
            print('H(',str(key),')=',str(self.H[key]))
        
        print('Delta = ',str(self.Delta))
    
    def create_Pi(self):
        #Description:
        # According to this paper, we need to define special set Pi.
        Pi = []
        for i in self.X:
            for j in self.X:
                #Key Criterion
                if self.H[i] is self.H[j]:
                    Pi += [ (i,j) ]

        self.Pi = Pi
        return Pi #For the sake of debugging
    
    def create_Theta(self):
        #Description:
        # According to this paper, we need to define special set Pi.
        Theta = []
        for i in self.X:
            Theta += [ (i,i) ]

        self.Theta = Theta
        return Theta #For the sake of debugging
    
    def succ(self,i):
        #Description:
        # According to paper
        #  succ(i) = { j in X : (i,j) in Delta }
        
        s_states = []
        for ij in self.Delta:
            if ij[0] is i:
                s_states += [ij[1]]
                
        return s_states
    
    def pre(self,i):
        #Description:
        # According to paper
        #   pre(i) = { j in X : (j,i) in Delta }
        
        p_states = []
        for ji in self.Delta:
            if ji[1] is i:
                p_states += [ji[0]]
                
        return p_states
    
    def S_kp1(self,S_k):
        #Description:
        # Equation (1) in the 2017 text describes a recursion that we would like to apply.
        #
        #Assumes:
        # - Pi already defined.
        temp_Skp1 = []
        temp_cp = []
        for ij in self.Pi:
            temp_cp = cart_prod(self.pre(ij[0]),self.pre(ij[1]))
            if intersection(temp_cp,S_k) is not None:
                temp_Skp1 += [ij]
            
        return temp_Skp1;
    
    def B_kp1(self,B_k):
        #Description:
        # To find the maximal set of pairs which are k-backward distinguishable for a given set, 
        # we will need to apply a tricky recursion for many values of k
        temp_Bkp1 = []
        for ij in B_k:
            #Note the formula written above.
            #1. compute the cartesian products of the two entries in the tuple ij
            temp_cp = cart_prod(self.pre(ij[0]),self.pre(ij[1]))
            #2. intersect cartesian product with B_k(\Sigma)
            temp_isect = intersection(temp_cp,B_k)
            #3. If intersection is nonempty, then add ij to B_{k+1}(\Sigma)
            if temp_isect:
                temp_Bkp1 += [ij]
                
        return temp_Bkp1
                
    
    def find_maximal_indist_set():
        print('Temp')

def diff(first, second):
        second = set(second)
        return [item for item in first if item not in second]

def cart_prod(l1,l2):
    #Description:
    # The cartesian product of the symbols listed in list 1 (l1) and list 2 (l2).
    
    product0 = []
    for item1 in l1:
        for item2 in l2:
            product0 += [(item1,item2)]
    
    return product0

def intersection(lst1,lst2):
    return [value for value in lst1 if value in lst2]

def set_minus(lst1,lst2):
    temp_sm = []
    for item1 in lst1:
        if not(item1 in lst2):
            temp_sm += [item1]
            
    return temp_sm
    
#Create the fsm version for diagnosability/indistinguishability analysis.
print('Considering Automatica FSM 2017 Paper.')
X5 = X2
X5_As = [X5[0],X5[2],X5[4]]
X5_Ms = diff(X5,X5_As)

X0_5 = X5[0];

Y5 = ['0','1']

H5 = dict()
for x in X5:
    if x in X5_As:
        H5[x] = Y5[1]
    else:
        H5[x] = Y5[0]
        
D5 = [ (X5[0],X5[1]),(X5[1],X5[2]),(X5[2],X5[3]),(X5[3],X5[4]),(X5[4],X5[0]) ]
        
A5 = fsm0(X5,X0_5,Y5,H5,D5)
print('Created "Automaton" (really an FSM) according to this paper.')
A5.print0()

print(' ')
print('Pi(A5)=',A5.create_Pi())
print('Theta(A5)=',A5.create_Theta())
print('S_{k+1} = ',A5.S_kp1( [(X0[0],X0[0])] ))
print(' ')
print('Testing Functions related to k-backward distinguishability.')
print('SET MINUS')
temp_l1 = [ 1 , 2 , 3 , 5]
temp_l2 = [ 3 , 4 , 5 ]
temp_l3 = [4,6,8]
temp_l4 = ['a1','a2','a3','a4']
temp_l5 = ['a2','a4']
print( 'set_minus(', str(temp_l1) , ',', str(temp_l2), ' ) = ', str(set_minus(temp_l1,temp_l2)) )
print( 'set_minus(', str(temp_l1) , ',', str(temp_l3), ' ) = ', str(set_minus(temp_l1,temp_l3)) )
print( 'set_minus(', str(temp_l4) , ',', str(temp_l5), ' ) = ', str(set_minus(temp_l4,temp_l5)) )
print(' ')
print('RECURSION')
print('A5.Pi\A5.Theta = ',set_minus(A5.Pi,A5.Theta))
print('B_{1}(A5.Pi\A5.Theta) = ', str(A5.B_kp1( set_minus(A5.Pi,A5.Theta) )))
print('B_{2}(A5.Pi\A5.Theta) = ', str(A5.B_kp1(A5.B_kp1( set_minus(A5.Pi,A5.Theta) ))))
print('B_{3}(A5.Pi\A5.Theta) = ', str(A5.B_kp1(A5.B_kp1(A5.B_kp1( set_minus(A5.Pi,A5.Theta) )))))

# print('B_2 = ',A5.B_kp1(  ))
```
